```python
edu_mode=[]
```

```python
job_types=car_train['Job'].value_counts().index
```

```python
job_types
```

```python
for job in job_types:
    mode=car_train[car_train['Job']==job]['Education'].value_counts().nlargest(1).index
    edu_mode=np.append(edu_mode,mode)
edu_map=pd.Series(edu_mode,index=job_types)
```

```python
edu_map
```

```python
for j in job_types:
    car_train.loc[(car_train['Education'].isnull())&(car_train['Job']==j),'Education']=edu_map.loc[edu_map.index==j][0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Separate numerical and categorical columns
numerical_cols = car_train.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = car_train.select_dtypes(include=['object']).columns

# Distribution plots for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(car_train[col], bins=20, kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Distribution plots for categorical columns
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=car_train[col], palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming car_train is your DataFrame
sns.boxplot(x=car_train['Age'])
plt.show()
```

```python
from scipy.stats import zscore

# Calculate z-scores
z_scores = zscore(car_train['Age'])

# Define a threshold for outliers (e.g., z-score > 3 or z-score < -3)
threshold = 3

# Find and print the indices of outliers
outlier_indices = (z_scores > threshold) | (z_scores < -threshold)
outliers = car_train['Age'][outlier_indices]
print("Outliers:", outliers)
```
